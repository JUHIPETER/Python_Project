```python
AddressData=pd.read_excel("Juhi_Exercise1.xlsx",sheet_name="Address Parsing")
```

```python
import re
from streetaddress import StreetAddressFormatter, StreetAddressParser
DF_Name= pd.DataFrame(AddressData)
DF_Name
address = DF_Name["full_address"]
len(address)
first=[]
second=[]
third=[]
forth=[]
fifth=[]
firstlist = []
secondlist = []
thirdlist = []
for i in range(0,len(address)):
    if("United States" in address[i]):
        firstlist.append(address[i])
    elif("France" in address[i]):
        secondlist.append(address[i])
    elif("Germany" in address[i]):
        thirdlist.append(address[i])
        
for s in firstlist:
    l =s.split(",")
    addr1 = l[0]
    city  = l[1]
    postcode = l[2]
    country = l[3]
    addr_parser = StreetAddressParser()
    addr = addr_parser.parse(addr1)
    address1=str(addr['house'])+" "+str(addr['street_full'])
    first.append(address1)
    second.append(addr['other'])
    third.append(city)
    forth.append(country)
    fifth.append(postcode)

for F in secondlist:
    try:
        l =F.split(",")
        address1 = l[0]
        address2 = " "
        addrsplit  = l[1]
        res = [re.findall(r'(\d+) (\w+)', addrsplit)]
        city = res[0][0][1]
        postcode = res[0][0][0]
        country = l[2]

    except:
        city="Na"
        postcode="Na"
    
    first.append(address1)
    second.append(address2)
    third.append(city)
    forth.append(country)
    fifth.append(postcode)

for G in thirdlist:
    try:
        l =G.split(",")

        address1 = l[0]
        address2  = " "
        addr2 = l[1]
        res = [re.findall(r'(\d+) (\w+)', addr2)]
        city = res[0][0][1]
        postcode = res[0][0][0]
        country = l[2]
    except:
        city="Na"
        postcode="Na"  
      
    first.append(address1)
    second.append(address2)
    third.append(city)
    forth.append(country)
    fifth.append(postcode)          
```

```python
import pandas as pd
df=pd.DataFrame(list(zip(first,second,third,forth,fifth)),columns=['Address1','Address2','City','Country','Postcode'])
```
